```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#load csv file data with headers

location = "datasets/DC_Properties.csv"
df = pd.read_csv(location)
```

```python
# See what the data looks like
df.head()
```

```python
# Statistical description of the data  
df.describe()
```

```python
# How many rows and columns?
df.shape
```

```python
# How many nulls are in each variable?
df.isnull().sum()
```

```python
# Are there duplicates in the addresses?
df.duplicated(['FULLADDRESS'])
```

```python
# Info on the duplicate addresses
dup_addy = df.duplicated(['FULLADDRESS'])
df.loc[dup_addy]
```

```python
# What are the actual addresses of the duplicates?
df[dup_addy]['FULLADDRESS']
```

```python
# What properties are less than $50,000?
df[df["PRICE"] < 50000]
```

```python
# What properties have more than 10 bathrooms and more than 10 rooms?
df[(df["BATHRM"] > 10) & (df["ROOMS"] > 10)]
```

```python
# What is the address of the properties with more than 10 bathrooms and more than 10 rooms?
df[(df["BATHRM"] > 10) & (df["ROOMS"] > 10)]["FULLADDRESS"]
```

```python
# In what quadrant of the city are the properties with more than 10 bathrooms and more than 10 rooms?
df[(df["BATHRM"] > 10) & (df["ROOMS"] > 10)]["QUADRANT"]
```

```python
# What is the average number of bathrooms in each quadrant of the city?
df.groupby('QUADRANT')['BATHRM'].mean()
```

```python
# What is the average number of bathrooms in houses in each year houses were built?
print(df.groupby('EYB')['BATHRM'].mean())
```
